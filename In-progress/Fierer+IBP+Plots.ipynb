```python
import pandas as pd
import os as os
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True)

TaxaTable = pd.read_table ("/Users/lorenayala/Desktop/Microbial/PhyllaDataset.txt")                           
print('Plots according to inundation factor.')
headers=list(TaxaTable.columns.values)
counter=2
i=0
for counter in range(1,10):
    for i in range (0,215):
        depth=TaxaTable.iloc[i,14]
        val=TaxaTable.iloc[i,counter]
        p=TaxaTable.iloc[i,17]
        
        if p >= 0 and p <0.1:
            x1=np.append([x1],[val])
            d1=np.append([d1],[depth])
        elif p >= 0.1 and p <0.2:
            x2=np.append([x2],[val])
            d2=np.append([d2],[depth])
        elif p >= 0.2 and p <0.3:
            x3=np.append([x3],[val])
            d3=np.append([d3],[depth])
        elif p >= 0.3 and p <0.4:
            x4=np.append([x4],[val])
            d4=np.append([d4],[depth])
        elif p >= 0.4 and p <0.5:
            x5=np.append([x5],[val])
            d5=np.append([d5],[depth])
        elif p >= 0.5 and p <0.6:
            x6=np.append([x6],[val])
            d6=np.append([d6],[depth])
        elif p >= 0.6 and p <0.7:
            x7=np.append([x7],[val])
            d7=np.append([d7],[depth])
        elif p >= 0.7 and p <0.8:
            x8=np.append([x8],[val])
            d8=np.append([d8],[depth])
        elif p >= 0.8 and p <0.9:
            x9=np.append([x9],[val])
            d9=np.append([d9],[depth])
        elif p >= 0.9 and p <=1:
            x10=np.append([x10],[val])
            d10=np.append([d10],[depth])
            continue
        else:
            continue

        i=i+1

    plt.scatter(x1,d1, label="0-10%",color="red")
    plt.scatter(x2,d2, label="10-20%",color="blue")
    plt.scatter(x3,d3, label="20-30%",color="green")
    plt.scatter(x4,d4, label="30-40%",color="yellow")
    plt.scatter(x5,d5, label="40-50%",color="orange")
    plt.scatter(x6,d6, label="50-60%",color="purple")
    plt.scatter(x7,d7, label="60-70%",color="black")
    plt.scatter(x8,d8, label="70-80%",color="brown")
    plt.scatter(x9,d9, label="80-90%",color="pink")
    plt.scatter(x10,d10, label="90-100%",color="gray")
    
    plt.legend(loc='lower right')
    plt.xlabel('Frequency')
    plt.ylabel('Depth (m)')
    plt.title(headers[counter])
    plt.ylim(1.2, 0)
    plt.xlim(0, 50000)
    plt.show()
    x1=x2=x3=x4=x5=x6=x7=x8=x9=x10=[]
    d1=d2=d3=d4=d5=d6=d7=d8=d9=d10=[]
    
    counter=counter+1
```
