```python
import numpy as np
import pandas as pd
import os as os

# Graphics
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
colormap = sns.color_palette('RdYlBu', 10)
%matplotlib inline
```

```python
# original paths
D = pd.read_table ("/Users/lorenayala/Desktop/Microbial/Phyla-Diversity-Files/Python-Abundance/Depth.txt") 
L = pd.read_table ("/Users/lorenayala/Desktop/Microbial/Phyla-Diversity-Files/Python-Abundance/Location.txt") 
I = pd.read_table ("/Users/lorenayala/Desktop/Microbial/Phyla-Diversity-Files/Python-Abundance/Inundation.txt") 
ST = pd.read_table ("/Users/lorenayala/Desktop/Microbial/Phyla-Diversity-Files/Python-Abundance/Soil Type.txt") 
```

```python
# relative paths for simplicity:
D = pd.read_table ("./data/Depth.txt") 
L = pd.read_table ("./data/Location.txt") 
I = pd.read_table ("./data/Inundation.txt")
ST = pd.read_table ("./data/Soil Type.txt") 
```

```python
headers=list(D.columns.values)

DN = np.arange(12)
LN = IN = np.arange(10)
STN = np.arange(6)
A=B=C=M=0
fig, axes = plt.subplots(ncols=2, nrows=2, sharey=True)
ax1, ax2, ax3, ax4 = axes.ravel()
ax1.set_ylabel('Relative Abundance',size=22)
ax1.set_xlabel('Depth (cm)',size=18)
ax2.set_xlabel('Well Number',size=18)
ax3.set_xlabel('Inundation Fraction (%)',size=18)
ax4.set_xlabel('Soil Type',size=18)
matplotlib.rc('ytick', labelsize=20)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18, 10)

for i in range (0,8):
    ax1.bar(DN, D.iloc[i,1:13], width=0.5,bottom=A,color=colormap[8-i],edgecolor="black",linewidth=1)
    ax2.bar(LN, L.iloc[i,1:11], width=0.5,bottom=B,color=colormap[8-i],edgecolor="black",linewidth=1)
    ax3.bar(IN, I.iloc[i,1:11], width=0.5,bottom=C,color=colormap[8-i],edgecolor="black",linewidth=1)
    ax4.bar(STN, ST.iloc[i,1:7], width=0.5,bottom=M,color=colormap[8-i],edgecolor="black",linewidth=1, label=ST.loc[i]['Phyla'])
    A=A+np.array(D.iloc[i,1:13])
    B=B+np.array(L.iloc[i,1:11])
    C=C+np.array(I.iloc[i,1:11])
    M=M+np.array(ST.iloc[i,1:7])
    i=i+1

# Add a legend:
handles, labels = ax4.get_legend_handles_labels()
# Need to add a label argument when calling the plotting function above, and for stacked bar charts need to reverse the order to match the chart. 
ax4.legend(handles[::-1], labels[::-1], title='Phyla',frameon=True, bbox_to_anchor=(1.25, .75))
    
plt.show()
```
