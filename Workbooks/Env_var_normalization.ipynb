```python
md_ph = pd.read_csv('../Metadata/Metadata_IBP_w_pH.txt', sep='\t')
md_3 = pd.read_csv('../Metadata/Metadata-IBP_3.txt', sep='\t')
md_ph2 = pd.read_csv('../Metadata/Metadata_IBP_w_pH_updated.txt')
```

```python
# Remove any added cells that got created during import. 

md_ph_columns = [x for x in md_ph.columns if 'Unnamed' not in x]
md_ph = md_ph[md_ph_columns]
md_ph = md_ph.iloc[:110,:]
md_ph.to_csv('../Metadata/Metadata_IBP_w_pH_updated.txt')
```

```python
# Add a label later to make matching sample names slightly easier. (still need to trim the replicte to match between the dataframes.)
def convert_name(well, section):
    name = 'W{0}.D{1}'.format(str(int(well)), str(int(section)))
    return name

md_ph2['sample_name'] = md_ph.apply(lambda row: convert_name(row['Well'], row['section ']), axis=1)
```

```python
# Fill missing values with mean for each column and then normalize the numeric columns.
md_ph2 = md_ph2.fillna(md_ph2.mean())
non_numeric = ['Well', 'section ', 'Soil type _color_id', 'Soil_type and color', 'USDA_Soil Series', 'sample_name']
numeric = [x for x in md_ph2.columns if x not in non_numeric]
md_normalized = md_ph2[numeric].apply(scipy.stats.zscore)
md_normalized[non_numeric] = md_ph2[non_numeric]
```

```python
new_df = pd.DataFrame(columns=md_normalized.columns)
for i,row in md_normalized.iterrows():
    new_df.append(row, ignore_index=True)
    new_df.loc[-1,'#SampleID'] = row['sample_name'] + '.rA'
    new_df.append(row, ignore_index=True)
    new_df.loc[i,'#SampleID'] 
```

```python
inds = ['A', 'B']
ind=0
with open('../Metadata/Metadata_IBP_w_pH_updated.txt', 'r') as f:
    with open('../Metadata/Metadata-IBP_w_ph_and_names.txt', 'w') as g:
        for line in f:
            pieces = line.split(',')
            if ind == 0:
                g.write(line)
            else:
                sampleID = 'W{0}.D{1}.{2}'.format(str(int(float(pieces[1]))), str(int(float((pieces[2])))), inds[0])
                sampleID2 = 'W{0}.D{1}.{2}'.format(str(int(float(pieces[1]))), str(int(float((pieces[2])))), inds[1])
                line_out1 = line.rstrip() + ',' + sampleID +'\n'
                line_out2 = line.rstrip() + ',' + sampleID2 +'\n'
                g.write(line_out1)
                g.write(line_out2)
            ind +=1
```

```python
# Transfer data from ibp_w_names_and_ph to the rows specified by the names in ibp-metadata
data_df = pd.read_csv('../Metadata/Metadata-IBP_w_ph_and_names.txt')
names_df = pd.read_csv('../Metadata/ibp-metadata.txt', sep='\t')


data_df_indexed = pd.DataFrame(index=names_df['#SampleID'], columns=data_cols)
for i, row in names_df.iterrows():
    if row['#SampleID'].startswith('W'):
        
        # Get data
        location = row['#SampleID'].split('.')
        well = location[0][1:]
        depth = location[1][1:]
        data_row = data_df.loc[(data_df['ind'] == float(well)) & (data_df['Well'] == float(depth))]

        # If data exists:
        if len(data_row) > 0:
            data = data_row.iloc[0][2:].values
            index = row['#SampleID']
            data_df_indexed.loc[index] = data
        
        
```
